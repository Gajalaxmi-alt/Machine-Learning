```python
# first step is to import pandas which is a library and it contains tables,classess and functions and recognised as a globlised one.
import pandas as pd 
```

```python
# making pandas to read the csv file and assigning the variables to dataset
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# when we want to find out what is in the dataset, it displays the Salary_Data.csv file with single input and output
dataset
```

```python
# to split the input column and output column , we are creating the input as independent and out put as dependant since the values are predicted based on input.
independent=dataset[["YearsExperience"]]
```

```python
# Input is the independent variable
independent
```

```python
dependent=dataset[["Salary"]]
```

```python
# ouyput is teh dependent variable
dependent
```

```python
# The next step is to split the train and test data, for this we are importing the functions of splititng  train and test data from sklearn model selection. This specific selection is due to the consumption of less memory. selecting teh parameters like independent, dependent, test valve as 70:30 ratio and random state =0 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# This is the learnign process phase
#to find out the weight and bias we ar importing the linear regression form sklearn linear model. these two lines are teh prodecure
# and the third line is the formula. The formula is executed with train sets.fit method is used with the subsitute x train and y train
# once the formula gets executed weight and bias will be created.
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# to find out the weight which is the slpoe
weight =regressor.coef_
weight
```

```python
# to find out the bias which is the origin
bias=regressor.intercept_
bias
```

```python
# model got craeted, in regressor weight, bias, model is created.
#evalaution metrics with teh test data
y_pred=regressor.predict(x_test)
```

```python
#to find out teh r2 value
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
# r2 to find out whether is a good model since teh vva;us is nearing to 1 the model has learnt well 
r_score
```

```python
#to save the model with a file name
import pickle
filename="finalized_model_linear.sav"
```

```python
# to run the model
pickle.dump(regressor, open(filename,'wb'))
```

```python
#to ;oad teh model and testing with the input
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
```
